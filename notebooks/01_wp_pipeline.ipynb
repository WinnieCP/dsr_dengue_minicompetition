```python
from sklearn.pipeline import Pipeline
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from sklearn.preprocessing import FunctionTransformer
```

```python
def select_features(df, features = None):
    # select features we want
    df_hlp = df[features].copy()
    return df_hlp

def deal_with_missing_values(df):
    # fill missing values with value from the past
    # we should probably make sure here that the dataframe
    # is ordered by time?
    df=df.fillna(method='ffill')
    return df

def select_city(df,city=None):
    # separate san juan and iquitos
    if 'city' in df.keys():
        df_city = df.loc[df.city == city]
        df_city = df_city.drop(columns = ['city'])
    else:
        df_city = df.loc[city]
    return df_city

myfeatures = ['reanalysis_specific_humidity_g_per_kg', 
                 'reanalysis_dew_point_temp_k', 
                 'station_avg_temp_c', 
                 'station_min_temp_c','week_start_date']

SelectCity = FunctionTransformer(func = select_city, check_inverse = False)

Impute = FunctionTransformer(func = deal_with_missing_values, check_inverse = False)

SelectFeatures = FunctionTransformer(func = select_features, check_inverse = False)
```

```python
pipe = Pipeline(steps=[('feature_selection', SelectFeatures), ('select_city', SelectCity), ('impute', Impute)])
```

```python
df_sj=pipe.set_params(feature_selection__kw_args={'features':myfeatures},
                      select_city__kw_args={'city':'sj'}
                     ).transform(df)
df_iq=pipe.set_params(feature_selection__kw_args={'features':myfeatures},
                      select_city__kw_args={'city':'iq'}
                     ).transform(df)
```
